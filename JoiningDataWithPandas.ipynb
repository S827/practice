## Right Join and Outer Join

```python
# if we want right join 2 tables, and we have a common column with same column names, 
# this is how we do it
new_table = gemini_1000.merge(gemini_5000, on="original_text", how='right')
new_table.head()
```

```python
# also we can add suffixes for the common columns on which we did not used to right join
new_table = gemini_1000.merge(gemini_5000, on='original_text', how='right', suffixes=('_1000', '_5000'))
new_table.head()
```

```python
# subset only the rows where reqrite_prompt_1000 is null
new_table_1 = new_table[new_table['rewrite_prompt_1000'].isnull()]
new_table_1.head()
```

```python
# if column names are not same on which we want to join 2 tables, but contains the same thing
# like id of table a and some_id of table b, contents are same but column names are different,
# in that case, we need to add 1 more argument, left_on='id', right_on='some_id'
```
